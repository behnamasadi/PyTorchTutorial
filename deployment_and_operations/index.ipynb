# **Deployment & Operations**

---

Once a model passes pre-deployment validation, *getting it into production safely and keeping it healthy* is its own discipline. This section covers how to ship a new model version with minimal risk, watch it in production, and prove that a new version is actually better than the old one.

The three concerns below map to the lifecycle of a deployed model: **how to release it**, **how to watch it**, and **how to measure it**.

## Contents

### 1. [Release & Rollout Strategies](release_rollout_strategies/index.ipynb)

How to push a new model version without a risky big-bang switch: blue-green deployments, canary releases, shadow/mirror traffic, and feature-flagged gradual rollouts — plus how to roll back fast when metrics regress.

### 2. [Production Monitoring](production_monitoring/index.ipynb)

Watching a live model: system metrics (latency, throughput, error rate, GPU utilization), model metrics (prediction distribution, confidence), and **data/concept drift** detection so you know *when* the model has gone stale.

### 3. [A/B Testing](ab_testing/index.ipynb)

Running a controlled experiment between model versions: traffic splitting, choosing a primary metric, statistical significance, and avoiding common pitfalls (peeking, sample-ratio mismatch) so the “winner” is real.

## How They Fit Together

```
  validated model
        │
        ▼
  [Release/Rollout]  ─ canary 5% → 50% → 100%, with instant rollback
        │
        ▼
  [Monitoring]       ─ latency, errors, drift  ─┐  feeds alerts & rollback
        │                                       │
        ▼                                       │
  [A/B Testing]      ─ is v2 actually better? ──┘
```

Rollout limits *blast radius*, monitoring *detects* problems, and A/B testing *decides* whether to keep the new version.
