# **Pre-Deployment Quality & Validation**

---

Before a model ever sees production traffic it must be **packaged into a portable, optimized artifact** and **gated by automated checks**. This section covers turning a training-time `nn.Module` into a fast, framework-independent inference artifact, and the quality bar it must clear first.

## Contents

### 1. [Pre-Release Quality Gates](pre_release_quality_gates/index.ipynb)

The automated checks a model must pass before release: accuracy/metric thresholds on a held-out set, regression tests against the current production model, latency and size budgets, and parity tests between the original and the exported model.

### 2. [Packaging for Inference](packaging_for_inference/index.ipynb)

Wrapping the trained model into a deployable artifact: stripping training-only code, pinning preprocessing, bundling weights + config, and producing a versioned, reproducible package (e.g. a container or model archive).

### 3. [Model Deployment: ONNX & TorchScript](model_deployment_ONNX_torchscript/index.ipynb)

Exporting a model out of eager PyTorch into a portable graph: **TorchScript** (`torch.jit.script`/`trace`) for Python-free C++/LibTorch serving, and **ONNX** (`torch.onnx.export`) as an interchange format runnable by ONNX Runtime, TensorRT, and others.

### 4. [TensorRT](tensor_rt.ipynb)

NVIDIA’s inference optimizer: compiling an ONNX/PyTorch graph into a hardware-specific engine with FP16/INT8 precision, kernel fusion, and layer auto-tuning for maximum GPU throughput and minimum latency.

## The Validation Pipeline

```
  trained nn.Module
        │
        ▼
  [Export]  ─ TorchScript / ONNX        (portable graph)
        │
        ▼
  [Optimize] ─ TensorRT FP16/INT8 engine (hardware-specific)
        │
        ▼
  [Package] ─ weights + preprocessing + config, versioned
        │
        ▼
  [Quality Gates] ─ accuracy, parity, latency, size → pass/fail
        │
        ▼
  ready for Deployment & Operations
```

Each export/optimize step can subtly change numerics, so **parity tests** (compare original vs. exported outputs within a tolerance) belong in the quality gates.
